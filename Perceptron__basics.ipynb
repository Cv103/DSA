```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#line y=x/2 divides data into two parts
#goal is to find weights that help separate data belonging to different class
a=np.array([[1.2,1],[1,2.1],[2.3,2],[2,1.5],[2,3],[-1.3,-1.1],[-1.7,-2.4],[-2,-2.1],[-2.3,-3]])
plt.scatter(a[::,:1:1],a[::,:0:-1])
x=np.arange(-2,2,0.01)
y=x/2
plt.plot(x,y)
```

```python

#one line seprates the data while other doesn't
x=np.arange(-3,3,0.01)
y=x
ax=x
ay=ax
plt.scatter(a[::,0:1:],a[::,1:2:])
plt.plot(x,y)
plt.plot(-ax,ay)
```

```python

#3 different lines trying to separate the data
x=np.array([[1,1],[1.2,2],[3,2],[3,-1],[-1,-3],[-2,-2.5],[-3,-1]])
y=np.array([1,1,1,0,0,0,0])
colormap=np.array(['r','g'])
arr=np.arange(-4,4,0.01)
ay=-(0.57)*arr-0.28
Ay=-(0.5)*arr
Aay=(0.22)*arr+0.44
ax = plt.gca()
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_position('zero')
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
plt.plot(arr,ay)
plt.plot(arr,Ay)
plt.plot(arr,Aay,label="correct boundry")
plt.legend()
plt.scatter(x[::,0:1:],x[::,1:2:],c=colormap [y])
```

```python
X = np.array([
        [0, 0],
        [0, 1],
        [1, 0],
        [1, 1]
    ])
Y = np.array([0, 1, 1, 1])
```

```python
#simple perceptron that learns OR operation

W=np.zeros(3)
K=[]
for i in range (10):
    for j in range (len(Y)):
        x=np.insert(X[j],0,1)
        u=W.T.dot(x)
        if u>=0:
            y=1
        else:
            y=0
        e=Y[j]-y
        K.append(W)
        W=W+0.33*e*x
```

```python
K=K[2:]
e
```

```python
#eq fails to seprate the data points
#sample plot of a eq given by learned weights
v=np.array(["r","g"])
l=np.arange(-10,12,0.01)
W=K[0]
W[-1]+=0.1
y=-(W[1]/W[2])*l-(W[0]/W[2])
print(y)
plt.plot(l,y,'r')
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.scatter(X[::,0:1:],X[::,1:2:], c=v[Y])
```

```python

#plot of decision boundry after each update; 
fig,ax=plt.subplots(4,4, figsize=(16,16))
p=0
for i in range(4):
    for j in range(4):
        W=K[p]
        p+=1
        l=np.arange(-4,4,0.01)
        y=-(W[1]/W[2])*l-(W[0]/W[2])
        ax[i,j].plot(l,y)
        ax[i,j].scatter(X[::,0:1:],X[::,1::], c=v[Y])
        ax[i,j].set_xlim(-2,3)
        ax[i,j].set_ylim(-2,3)
```

```python
x=np.array([[1,1],[1.2,2],[3,2],[3,-1],[-1,-3],[-2,-2.5],[-3,-1]])
y=np.array([1,1,1,0,0,0,0])
cmap=y
colormap=np.array(['r','g'])
plt.scatter(x[::,0:1:],x[::,1:2:],c= colormap [y])
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("data scatter plot")
plt.text(-1.3,-0.5,"linearly seprable data",fontsize=12,color='g',bbox={'facecolor': 'red', 'alpha': 0.5, 'pad': 10})
```

```python
W=np.zeros(3)
error=[]
K=[]
epoch=19
lr=0.02
for i in range (epoch):
    for j in range (len(x)):
        X=np.insert(x[j],0,1)
        u=W.T.dot(X)
        if u>=0:
            Y=1
        else:
            Y=0
        e=y[j]-Y
        error.append(e)
        W=W+lr*e*X
        K.append(W)
```

```python
plt.plot(list(range (len(error))),error,label="error",color='r')
plt.xlim(-1,25)
plt.ylim(-4,4)
plt.legend()
```

```python
fig,ax=plt.subplots(2,2, figsize=(16,16))
p=4
for i in range(2):
    for j in range(2):
        W=K[p]
        p+=1
        l=np.arange(-4,4,0.01)
        y=-(W[1]/W[2])*l-(W[0]/W[2])
        ax[i,j].plot(l,y, label='boundry',color='purple')
        ax[i,j].scatter(x[::,0:1:],x[::,1::], c=v[cmap])
        ax[i,j].set_xlim(-6,6)
        ax[i,j].set_ylim(-6,6)
        ax[i,j].legend()
```
